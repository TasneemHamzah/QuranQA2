```python
from keras.layers import LSTM, Dropout, Dense, Embedding, Input
from keras import Sequential
from keras.models import Model
from keras.layers.convolutional import Conv1D, MaxPooling1D
from keras.layers.merge import concatenate

vocab_size=12000
length=200
model = Sequential([
Embedding(vocab_size, 50, input_length= length),
#Dropout(0.2),
LSTM(100, return_sequences=True),
LSTM(100, return_sequences=True),
LSTM(100),
Dropout(0.2),
Dense(512,activation='relu' ),
Dropout(0.2),
Dense(256 ,activation='relu' ),
Dense(vocab_size, activation='softmax')
])
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.summary()
```

```python
from keras.utils.vis_utils import plot_model

plot_model(model, 'NLP.jpg', show_shapes=True)
```

```python
from keras.layers import LSTM, Dropout, Dense, Embedding, Input, Flatten
from keras import Sequential
from keras.models import Model
from keras.layers.convolutional import Conv1D, MaxPooling1D
from keras.layers.merge import concatenate

def build_model(vocab_size, length):
    inputs1 = Input(shape=(length,))
    embedding1= Embedding(vocab_size, 50)(inputs1)
    #lstm = LSTM(128, return_sequences=True)(embedding)
    #drop = Dropoyt(0,2)(lstm)
    #lstm = LSTM(128)(drop)
    conv1 = Conv1D(filters=32, kernel_size=4, activation='relu')(embedding1)
    drop1 = Dropout(0,2)(conv1)
    pool1=MaxPooling1D(pool_size=2)(drop1)
    flat1= Flatten()(pool1)
    
    inputs2 = Input(shape=(length,))
    embedding2= Embedding(vocab_size, 50)(inputs2)
    #lstm = LSTM(128, return_sequences=True)(embedding)
    #drop = Dropoyt(0,2)(lstm)
    #lstm = LSTM(128)(drop)
    conv2 = Conv1D(filters=32, kernel_size=4, activation='relu')(embedding2)
    drop2 = Dropout(0,2)(conv2)
    pool2=MaxPooling1D(pool_size=2)(drop2)
    flat2= Flatten()(pool2)
    
    inputs3 = Input(shape=(length,))
    embedding3= Embedding(vocab_size, 50)(inputs3)
    #lstm = LSTM(128, return_sequences=True)(embedding)
    #drop = Dropoyt(0,2)(lstm)
    #lstm = LSTM(128)(drop)
    conv3 = Conv1D(filters=32, kernel_size=4, activation='relu')(embedding3)
    drop3 = Dropout(0,2)(conv3)
    pool3=MaxPooling1D(pool_size=2)(drop3)
    flat3= Flatten()(pool3)
    
    channels = concatenate([flat1,flat2,flat3])
    
    dense = Dense(512,activation='relu' )(channels)
    outputs= Dense(vocab_size, activation='softmax')(dense)
    model = Model(inputs=[inputs1,inputs2,inputs3], outputs= outputs)
    
    return model

model = build_model(1200, 50)
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.summary()
```

```python
from keras.utils.vis_utils import plot_model

plot_model(model, 'NLP2.jpg', show_shapes=True)
```
